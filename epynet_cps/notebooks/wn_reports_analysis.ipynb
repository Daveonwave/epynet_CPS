```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

to_load = 'to_load/'
```

```python
def plot_report_data(filename, entity_type, uids, prop):
    """
    Plot data retrieve from links or nodes csv report.
    :param filename: name of the csv report file
    :param entity_type: kind of links or nodes to plot
    :param uids: ids of the entities to plot
    :param prop: variable to plot
    """
    df = pd.read_csv(filename, header=[0, 1, 2])
    
    fig, axs = plt.subplots(1, 1, figsize=(15,7), sharex=True)
    for uid in uids:
        axs.plot(df[entity_type][uid][prop], label=uid)
    axs.set_title(entity_type + ': ' + prop)
    axs.legend()
```

```python
csv_file = "trial_pump_status_issue/links0.csv"
filename = to_load + csv_file

entity_type = 'pumps'
uids = ['P78', 'P79']
prop = 'energy'
plot_report_data(filename=filename, entity_type=entity_type, uids=uids, prop=prop)
```
